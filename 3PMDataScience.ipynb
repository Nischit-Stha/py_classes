Feature Engineering
Feature engineering involves transforming raw data into meaningful features that improve model performance. This may involve:

1. Creating new features: Based on domain knowledge or transformations (e.g., combining two columns into one).
2. Encoding categorical variables: Using techniques like one-hot encoding or label encoding.
3. Scaling features: Standardization (zero mean, unit variance) or Min-Max scaling (to a fixed range).

````python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
import pandas as pd
import numpy as np

# Example dataset with categorical and numerical features
data = pd.DataFrame({
    'age': [25, 30, np.nan, 45],
    'salary': [50000, 60000, 70000, 80000],
    'gender': ['M', 'F', 'M', 'F']
})

# Define preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', Pipeline([
            ('imputer', SimpleImputer(strategy='mean')),
            ('scaler', StandardScaler())
        ]), ['age', 'salary']),
        ('cat', OneHotEncoder(), ['gender'])
    ]
)

# Apply transformations
data_transformed = preprocessor.fit_transform(data)
print(data_transformed)
````

**Statistical Inference and Probability Distributions**

Statistical inference involves making conclusions about populations based on sample data.

Probability distributions describe the likelihood of different outcomes in an uncertain process.

Common probability distributions:

1. Normal Distribution: Bell-shaped curve, used for continuous data.
2. Binomial Distribution: Discrete, used for binary outcomes.
3. Poisson Distribution: Describes the number of events occurring in a fixed interval of time.

Hypothesis Testing
Hypothesis testing is used to validate assumptions or claims about a population. The null hypothesis (H₀) represents a default position (e.g., no effect), and the alternative hypothesis (H₁) represents the claim to test. Common tests include:

1. t-test: Compares means between two groups.
2. Chi-square test: Compares categorical data.
3. ANOVA: Compares means across multiple groups.

````python
from scipy import stats

# Sample data
group1 = [20, 22, 23, 25, 27]
group2 = [30, 31, 29, 35, 33]
# group1 = [75, 80, 85, 90, 95]  # Group 1: Higher average
# group2 = [60, 65, 70, 80, 85]  # Group 2: Lower average

# Perform a t-test
t_stat, p_value = stats.ttest_ind(group1, group2)

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis - significant difference between groups")
else:
    print("Fail to reject the null hypothesis - no significant difference")
````

1. If the p-value is < 0.05, reject the null hypothesis (there's a significant effect).
2. If the p-value is > 0.05, don't reject the null hypothesis (no significant effect).

````python
from scipy.stats import chi2_contingency
import numpy as np

# Example data: gender vs. product preference
# Rows represent gender (Male, Female), and columns represent product preferences (Product A, Product B)
data = np.array([[30, 10],  # 30 males prefer Product A, 10 males prefer Product B
                 [25, 15]])  # 25 females prefer Product A, 15 females prefer Product B

# Perform chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(data)

# Output the results
print(f"Chi-square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies: \n{expected}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis - there is an association between gender and product preference")
else:
    print("Fail to reject the null hypothesis - no association between gender and product preference")
````
